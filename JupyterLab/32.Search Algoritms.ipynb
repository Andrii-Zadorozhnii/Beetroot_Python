üîπ 1. –ü—É–∑—ã—Ä—å–∫–æ–≤–∞—è —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ (Bubble Sort)

–ò–¥–µ—è: –ü—Ä–æ—Ö–æ–¥–∏–º –ø–æ —Å–ø–∏—Å–∫—É, —Å—Ä–∞–≤–Ω–∏–≤–∞–µ–º —Å–æ—Å–µ–¥–Ω–∏–µ —ç–ª–µ–º–µ–Ω—Ç—ã –∏ –º–µ–Ω—è–µ–º –∏—Ö –º–µ—Å—Ç–∞–º–∏, –µ—Å–ª–∏ –æ–Ω–∏ —Å—Ç–æ—è—Ç –Ω–µ –≤ —Ç–æ–º –ø–æ—Ä—è–¥–∫–µ. –ü–æ–≤—Ç–æ—Ä—è–µ–º, –ø–æ–∫–∞ –º–∞—Å—Å–∏–≤ –Ω–µ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω.

üîª –°–ª–æ–∂–Ω–æ—Å—Ç—å: O(n¬≤) (–ø–ª–æ—Ö–æ –¥–ª—è –±–æ–ª—å—à–∏—Ö –º–∞—Å—Å–∏–≤–æ–≤)
‚úÖ –ü—Ä–æ—Å—Ç–∞—è —Ä–µ–∞–ª–∏–∑–∞—Ü–∏—è


   
    –ü—É–∑—ã—Ä—å–∫–æ–≤–∞—è —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞. –ü–æ–≤—Ç–æ—Ä—è–µ–º –ø—Ä–æ—Ö–æ–¥—ã, –ø–æ–∫–∞ –º–∞—Å—Å–∏–≤ –Ω–µ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω.
    def bubble_sort(arr: list[int]) -> list[int]:
        """
        :param arr: –ò—Å—Ö–æ–¥–Ω—ã–π —Å–ø–∏—Å–æ–∫ —á–∏—Å–µ–ª.
        :return: –û—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫.
        """
        n = len(arr)
        for i in range(n):
            swapped = False  # –§–ª–∞–≥ –¥–ª—è –æ–ø—Ç–∏–º–∏–∑–∞—Ü–∏–∏ (–µ—Å–ª–∏ –Ω–µ –±—ã–ª–æ –æ–±–º–µ–Ω–æ–≤ - –º–∞—Å—Å–∏–≤ —É–∂–µ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω)
            for j in range(n - i - 1):
                if arr[j] > arr[j + 1]:
                    arr[j], arr[j + 1] = arr[j + 1], arr[j]  # –û–±–º–µ–Ω –º–µ—Å—Ç–∞–º–∏
                    swapped = True
            if not swapped:
                break
    return arr

    )



üîπ 2. –°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –≤—Å—Ç–∞–≤–∫–∞–º–∏ (Insertion Sort)

–ò–¥–µ—è: –ë–µ—Ä—ë–º —ç–ª–µ–º–µ–Ω—Ç –∏ –≤—Å—Ç–∞–≤–ª—è–µ–º –µ–≥–æ –≤ —É–∂–µ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—É—é —á–∞—Å—Ç—å –º–∞—Å—Å–∏–≤–∞.

üîª –°–ª–æ–∂–Ω–æ—Å—Ç—å: O(n¬≤) (–º–µ–¥–ª–µ–Ω–Ω–æ –Ω–∞ –±–æ–ª—å—à–∏—Ö –¥–∞–Ω–Ω—ã—Ö)
‚úÖ –•–æ—Ä–æ—à –¥–ª—è –ø–æ—á—Ç–∏ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö (O(n) –≤ –ª—É—á—à–µ–º —Å–ª—É—á–∞–µ)

def insertion_sort(arr: list[int]) -> list[int]:
    """
    –°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –≤—Å—Ç–∞–≤–∫–∞–º–∏. –ë–µ—Ä—ë–º —ç–ª–µ–º–µ–Ω—Ç –∏ –≤—Å—Ç–∞–≤–ª—è–µ–º –µ–≥–æ –≤ –ø—Ä–∞–≤–∏–ª—å–Ω–æ–µ –º–µ—Å—Ç–æ –≤ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–π —á–∞—Å—Ç–∏ —Å–ø–∏—Å–∫–∞.
    
    :param arr: –ò—Å—Ö–æ–¥–Ω—ã–π —Å–ø–∏—Å–æ–∫ —á–∏—Å–µ–ª.
    :return: –û—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫.
    """
    for i in range(1, len(arr)):
        key = arr[i]  # –ó–∞–ø–æ–º–∏–Ω–∞–µ–º —Ç–µ–∫—É—â–∏–π —ç–ª–µ–º–µ–Ω—Ç
        j = i - 1
        while j >= 0 and arr[j] > key:  # –î–≤–∏–≥–∞–µ–º —ç–ª–µ–º–µ–Ω—Ç—ã –≤–ø—Ä–∞–≤–æ, –ø–æ–∫–∞ –Ω–µ –Ω–∞–π–¥—ë–º –º–µ—Å—Ç–æ
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key  # –í—Å—Ç–∞–≤–ª—è–µ–º —ç–ª–µ–º–µ–Ω—Ç –Ω–∞ –Ω—É–∂–Ω–æ–µ –º–µ—Å—Ç–æ
    return arr

numbers = [5, 2, 9, 1, 5, 6]

)

üîπ 3. –°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –≤—ã–±–æ—Ä–æ–º (Selection Sort)

–ò–¥–µ—è: –ù–∞—Ö–æ–¥–∏–º –º–∏–Ω–∏–º—É–º –∏ —Å—Ç–∞–≤–∏–º –µ–≥–æ –≤ –Ω–∞—á–∞–ª–æ, –∑–∞—Ç–µ–º –∏—â–µ–º —Å–ª–µ–¥—É—é—â–∏–π –º–∏–Ω–∏–º—É–º –∏ —Ç. –¥.

üîª –°–ª–æ–∂–Ω–æ—Å—Ç—å: O(n¬≤)
‚úÖ –ü—Ä–æ—Å—Ç–æ–π –∞–ª–≥–æ—Ä–∏—Ç–º, –Ω–æ –º–µ–¥–ª–µ–Ω–Ω—ã–π

def selection_sort(arr: list[int]) -> list[int]:
    """
    –°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –≤—ã–±–æ—Ä–æ–º. –ù–∞—Ö–æ–¥–∏–º –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–π —ç–ª–µ–º–µ–Ω—Ç –∏ —Å—Ç–∞–≤–∏–º –µ–≥–æ –≤ –Ω–∞—á–∞–ª–æ –º–∞—Å—Å–∏–≤–∞.
    
    :param arr: –ò—Å—Ö–æ–¥–Ω—ã–π —Å–ø–∏—Å–æ–∫ —á–∏—Å–µ–ª.
    :return: –û—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫.
    """
    n = len(arr)
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]  # –ú–µ–Ω—è–µ–º –º–µ—Å—Ç–∞–º–∏
    return arr

üîπ 4. –ë—ã—Å—Ç—Ä–∞—è —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ (QuickSort)

–ò–¥–µ—è: –í—ã–±–∏—Ä–∞–µ–º ‚Äú–æ–ø–æ—Ä–Ω—ã–π‚Äù —ç–ª–µ–º–µ–Ω—Ç (pivot), –¥–µ–ª–∏–º –º–∞—Å—Å–∏–≤ –Ω–∞ —Ç—Ä–∏ —á–∞—Å—Ç–∏:
	‚Ä¢	—ç–ª–µ–º–µ–Ω—Ç—ã –º–µ–Ω—å—à–µ pivot,
	‚Ä¢	—ç–ª–µ–º–µ–Ω—Ç—ã —Ä–∞–≤–Ω—ã–µ pivot,
	‚Ä¢	—ç–ª–µ–º–µ–Ω—Ç—ã –±–æ–ª—å—à–µ pivot.
–†–µ–∫—É—Ä—Å–∏–≤–Ω–æ —Å–æ—Ä—Ç–∏—Ä—É–µ–º –ø–æ–¥–º–∞—Å—Å–∏–≤—ã.

üîπ –°–ª–æ–∂–Ω–æ—Å—Ç—å: O(n log n) (–≤ —Å—Ä–µ–¥–Ω–µ–º), O(n¬≤) (–≤ —Ö—É–¥—à–µ–º —Å–ª—É—á–∞–µ)
‚úÖ –û—á–µ–Ω—å –±—ã—Å—Ç—Ä—ã–π –∞–ª–≥–æ—Ä–∏—Ç–º –¥–ª—è –±–æ–ª—å—à–∏–Ω—Å—Ç–≤–∞ —Å–ª—É—á–∞–µ–≤

def quick_sort(arr: list[int]) -> list[int]:
    """
    –ë—ã—Å—Ç—Ä–∞—è —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞. –í—ã–±–∏—Ä–∞–µ–º –æ–ø–æ—Ä–Ω—ã–π —ç–ª–µ–º–µ–Ω—Ç, –¥–µ–ª–∏–º —Å–ø–∏—Å–æ–∫ –Ω–∞ —á–∞—Å—Ç–∏ –∏ —Å–æ—Ä—Ç–∏—Ä—É–µ–º —Ä–µ–∫—É—Ä—Å–∏–≤–Ω–æ.
    
    :param arr: –ò—Å—Ö–æ–¥–Ω—ã–π —Å–ø–∏—Å–æ–∫ —á–∏—Å–µ–ª.
    :return: –û—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫.
    """
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]  # –û–ø–æ—Ä–Ω—ã–π —ç–ª–µ–º–µ–Ω—Ç
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

üîπ 5. –°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ —Å–ª–∏—è–Ω–∏–µ–º (Merge Sort)

–ò–¥–µ—è: –†–∞–∑–±–∏–≤–∞–µ–º –º–∞—Å—Å–∏–≤ –Ω–∞ –¥–≤–µ —á–∞—Å—Ç–∏, —Å–æ—Ä—Ç–∏—Ä—É–µ–º –∫–∞–∂–¥—É—é —Ä–µ–∫—É—Ä—Å–∏–≤–Ω–æ –∏ —Å–ª–∏–≤–∞–µ–º –∏—Ö.

üîπ –°–ª–æ–∂–Ω–æ—Å—Ç—å: O(n log n)
‚úÖ –°—Ç–∞–±–∏–ª—å–Ω–∞—è —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞, –ø–æ–¥—Ö–æ–¥–∏—Ç –¥–ª—è LinkedList

def merge_sort(arr: list[int]) -> list[int]:
    """
    –°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ —Å–ª–∏—è–Ω–∏–µ–º. –†–∞–∑–¥–µ–ª—è–µ–º –º–∞—Å—Å–∏–≤ –Ω–∞ —á–∞—Å—Ç–∏, —Å–æ—Ä—Ç–∏—Ä—É–µ–º –∏ –æ–±—ä–µ–¥–∏–Ω—è–µ–º –æ–±—Ä–∞—Ç–Ω–æ.
    
    :param arr: –ò—Å—Ö–æ–¥–Ω—ã–π —Å–ø–∏—Å–æ–∫ —á–∏—Å–µ–ª.
    :return: –û—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫.
    """
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])

    # –§—É–Ω–∫—Ü–∏—è –¥–ª—è –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏—è –¥–≤—É—Ö –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö —Å–ø–∏—Å–∫–æ–≤
    def merge(left: list[int], right: list[int]) -> list[int]:
        result = []
        i = j = 0
        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                result.append(left[i])
                i += 1
            else:
                result.append(right[j])
                j += 1
        result.extend(left[i:])
        result.extend(right[j:])
        return result

    return merge(left, right)
