Список (list) — это изменяемый, упорядоченный набор данных, который может содержать элементы любых типов. Списки создаются с использованием квадратных скобок [].

```python
# Списки в Python

# Создание списков
empty_list = []  # Пустой список
numbers = [1, 2, 3, 4, 5]  # Список с элементами
mixed_list = [1, "hello", 3.14, [1, 2]]  # Список с элементами разных типов

# Основные операции
# 1. Индексация
print(numbers[0])  # Доступ по индексу (первый элемент)
print(numbers[-1])  # Доступ к последнему элементу

# 2. Изменение элемента
numbers[1] = 25
print(numbers)

# 3. Срезы
print(numbers[1:3])  # Доступ к подмножеству элементов
print(numbers[:2])  # Элементы от начала до индекса 2

# 4. Длина списка
print(len(numbers))  # Количество элементов в списке

# 5. Проверка наличия элемента
print(30 in numbers)

# Методы списков
numbers = [10, 20, 30, 20]

# append(x) - Добавление элемента в конец
numbers.append(40)
print(numbers)

# extend(iterable) - Расширение списка другим списком
numbers.extend([50, 60])
print(numbers)

# insert(i, x) - Вставка элемента на позицию i
numbers.insert(2, 15)
print(numbers)

# remove(x) - Удаление первого найденного элемента
numbers.remove(20)
print(numbers)

# pop([i]) - Удаление элемента по индексу (по умолчанию последний)
last = numbers.pop()
print(numbers, last)

# clear() - Очистка списка
numbers.clear()
print(numbers)

# index(x) - Возвращает индекс элемента
numbers = [10, 20, 30, 20]
print(numbers.index(20))

# count(x) - Количество вхождений элемента
print(numbers.count(20))

# sort(key=None, reverse=False) - Сортировка списка
numbers = [4, 2, 8, 1]
numbers.sort()  # По возрастанию
print(numbers)
numbers.sort(reverse=True)  # По убыванию
print(numbers)

# reverse() - Разворот списка
numbers.reverse()
print(numbers)

# copy() - Создание копии списка
new_numbers = numbers.copy()
print(new_numbers)

# Итерация по списку
numbers = [1, 2, 3, 4]
for num in numbers:
    print(num)

# Генераторы списков
squares = [x**2 for x in range(5)]
print(squares)

# Вложенные списки
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# Доступ к элементам вложенных списков
print(matrix[1][2])
```
